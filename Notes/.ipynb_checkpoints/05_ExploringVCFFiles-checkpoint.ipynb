<div class="alert alert-block alert-warning">
    <h3>2.3 Missingness per individual?</h3>
    <p>VCF files often denote a missing variants with a <b>./.</b> symbol. Too much missingness is concerning, and so is no missingness at all (is the VCF file pruned?)</p>
</div>

```python
#Check missingness.
!~/biol6150/SharedEnvironment/bin/vcftools --vcf ~/biol6150/Data/1000Genomes.Pruned.vcf --missing-indv --out ~/scratch/VCFExploration/1000Genomes.missing
```

```python
#Check missingness.
!cat ~/scratch/VCFExploration/1000Genomes.missing.imiss | head
```

<div class="alert alert-block alert-warning">
    <h3>2.4 Allele frequencies</h3>
    <p>Allele frequencies are important metric for variants in clinical genomics, population & evolutionary genetics</p>
</div>

```python
#Calculate allele frequencies.
!~/biol6150/SharedEnvironment/bin/vcftools --vcf ~/biol6150/Data/1000Genomes.Pruned.vcf --freq --out ~/scratch/VCFExploration/1000Genomes.frequencies
```

```python
#Check frequencies file.
!cat ~/scratch/VCFExploration/1000Genomes.frequencies.frq | head
```

```python
#Check the WRAP53 VCF file.
!cat ~/scratch/VCFExploration/1000Genomes.WRAP53.recode.vcf | head -n30 | cut -f1-12
```

<div class="alert alert-block alert-warning">
    <h3>2.6 I'm interested in the dbSNP variant rs76560455, how can I get information for just that variant?</h3>
    <p>Check how variants are coded in your VCF file (what does the ID column represent; can you recode it?)</p>
</div>

```python
#Check the source VCF file first
!head ~/biol6150/Data/1000Genomes.Pruned.vcf -n30 | cut -f1-12
```

```python
#Get variants that lie on WRAP53 gene.
!~/biol6150/SharedEnvironment/bin/vcftools --vcf ~/biol6150/Data/1000Genomes.Pruned.vcf --snp chr2:188658150:T:C --recode --out ~/scratch/VCFExploration/1000Genomes.rs76560455
```

```python
#Check the VCF file.
!cat ~/scratch/VCFExploration/1000Genomes.rs76560455.recode.vcf | cut -f1-12
```

<div class="alert alert-block alert-warning">
    <h3>2.7 I want all CHB individuals.</h3>
    <p>Check how samples are coded in your VCF file. Make a desired samples' file</p>
</div>

```python
#Check how samples are coded.
!cat ~/scratch/VCFExploration/1000Genomes.rs76560455.recode.vcf | grep "#CHR" | tr "\t" "\n" | grep "CHB" | head
```

```python
#Get all CHB samples. 
!cat ~/scratch/VCFExploration/1000Genomes.rs76560455.recode.vcf | grep "#CHR" | tr "\t" "\n" | grep "CHB" > ~/scratch/VCFExploration/1000Genomes.CHB.samples
```

```python
#Get samples that belong to CHB family gene.
!~/biol6150/SharedEnvironment/bin/vcftools --vcf ~/biol6150/Data/1000Genomes.Pruned.vcf --keep ~/scratch/VCFExploration/1000Genomes.CHB.samples --recode --out ~/scratch/VCFExploration/1000Genomes.CHB
```

```python
#Check the VCF file.
!cat ~/scratch/VCFExploration/1000Genomes.CHB.recode.vcf | cut -f1-12 | head -n40
```

<div class="alert alert-block alert-warning">
    <h3>2.8 Look at SNP density.</h3>
</div>

```python
#Get density.
!~/biol6150/SharedEnvironment/bin/vcftools --vcf ~/biol6150/Data/1000Genomes.Pruned.vcf --chr 1 --SNPdensity 100000 --out ~/scratch/VCFExploration/1000Genomes.Density
```

```python
#Check the VCF file density.
!cat ~/scratch/VCFExploration/1000Genomes.Density.snpden | head
```
