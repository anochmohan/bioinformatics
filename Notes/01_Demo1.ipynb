## 1. Basic jupyter functionality

### 1.1 Basics of jupyter cell types

The greatness of jupyter lies in having individual cells, and being able to run and analyze the results of cells individually. This makes it really easy to show other people your thought process, document your work, show what you tried --- in other words, reproducible bioinformatics.

By default, cells in jupyter are "Code" cells. To convert a cell to markdown, which is the cell type you are reading right now, go to _Cell > Cell Type > Markdown_.
\
\
As a shortcut, hit _esc_ on your keyboard or click the whitespace to the left of a cell and then press "__m__" on your keyboard to switch the current cell to markdown.  
\
\
To run any type of cell, hit __Shift+Enter__. 
\
\
Here is a helpful resource for text formatting in jupyter markdown: https://www.ibm.com/docs/en/watson-studio-local/1.2.3?topic=notebooks-markdown-jupyter-cheatsheet


### 1.2 Adding and deleting cells

To add new cells, you can go to _Insert > Insert Cell Above_ to insert a cell above your current cell, or go to _Insert > Insert Cell Below_ to insert a cell below you current cell. 
\
\
As a shortcut, hit _esc_ on your keyboard or click the whitespace to the left of a cell and then press "__a__" to add a cell _above_ your current cell, or press "__b__" to add a cell _below_ your current cell (an easy way to remember this is "A" for "Above" and "B" for "Below").
\
\
Go ahead and try adding a few cells above and below this cell.

You can also delete cells by hitting escape and then double tapping _d_.

## 3. Sample data analysis using Python libraries

You can import Python modules directory into a jupyter notebook. A python module is a package containing pre-written programs and functions that you can call and apply for your convenience. 

```python
from pylab import randn
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

### 3.1 Generate a simple plot using matplotlib.

```python
# Generate 500 random numbers and plot X vs Y. 
X = randn(500)
Y = randn(500)
plt.scatter(X,Y)

# Add x and y axis labels. 
plt.xlabel("Variable X")
plt.ylabel("Variable Y")

plt.show()
```

There are many different ways you can customize a plot to make it publciation-quality, increase readibility, make it more aesthetically appealing, or make a plot specifically for _your_ analysis. The documentations for each of the python modules is available online, and Google (or ChatGPT, or Bard AI, etc.) is your friend. 
\
\
matplotlib documentation:
\
\
Here is the sample plot, made a bit more publication-quality.

```python
%matplotlib inline
import seaborn as sns 
custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="ticks", rc=custom_params)

# Generate 500 random numbers and plot X vs Y. 
X = randn(500)
Y = randn(500)
plt.scatter(X,Y, color='royalblue', edgecolor='black')

# Set axis limits 
plt.xlim(-4,4)
plt.ylim(-4,4)

# Set the font size for the tick labels
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

# Add x and y axis labels. 
plt.title('Plotting X vs Y',size=20)
plt.xlabel('Variable X', size=18)
plt.ylabel('Variable Y', size=18)

## Un-comment the lines below to save the plot as a png file in your working directory. 
## Shortcut to un-comment in jupyter: highlight lines with your mouse,
## then hit command+"/" (Mac) or alt+"/" (windows)

mydir = !pwd
plt.tight_layout()
plt.savefig('{}/sample_fig.pdf'.format(mydir[0]), dpi=300)

plt.show()
```

 If you decided to save as a `png`, confirm `sample_fig.png` was generated by running `ls -lt`

### 3.2 Make the plot more suitable to see.

```python
plt.scatter(X,Y, color='royalblue')

# Set axis limits 
plt.xlim(-4,4)
plt.ylim(-4,4)

# Set the font size for the tick labels
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

# Add x and y axis labels. 
plt.title('Plotting X vs Y',size=20)
plt.xlabel('Variable X', size=18)
plt.ylabel('Variable Y', size=18)

plt.rcParams['figure.figsize'] = [24, 24]

plt.show()
```
